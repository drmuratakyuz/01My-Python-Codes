```python
for i in range(10,0,-1):
  for j in range(1,4):
    print(i,j)
```

```python
for a in range(8):
  for i in range(10,0,-1):
    for j in range(1,4):
      print(a,i,j)
```

```python
def kidem_derece(ilk,yil):  
  for a in range(0,yil):
    for i in range(ilk,0,-1):
      for j in range(1,4):
        print(i,j)

kidem_derece(15, 8)     
```
